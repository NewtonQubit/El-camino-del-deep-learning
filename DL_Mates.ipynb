```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def f(x):
    return x**2

def grad_f(x):
    return 2 * x

# Parámetros iniciales
x = 10.0  # Valor inicial, numero aleatorio
learning_rate = 0.01
num_iterations = 500
x_values = [x]
f_values = [f(x)]

# Algoritmo de Descenso de Gradiente
for i in range(num_iterations):
    grad = grad_f(x)
    x = x - learning_rate * grad
    x_values.append(x)
    f_values.append(f(x))

# Visualización
x_plot = np.linspace(-10, 10, 400)
f_plot = f(x_plot)

plt.figure(figsize=(8, 6))
plt.plot(x_plot, f_plot, label='$f(x) = x^2$')
plt.plot(x_values, f_values, 'ro-', label='Descenso de Gradiente')
plt.title('Descenso de Gradiente sobre $f(x) = x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def f(x):
    return  x**4 - 3*x**3 + x

def grad_f(x):
    return 4*x**3 - 3*3*x**2 + 1

# Parámetros iniciales
x = -1.1  # Inicial cerca del mínimo local
learning_rate = 0.01
num_iterations = 500
x_values = [x]
f_values = [f(x)]

# Algoritmo de Descenso de Gradiente
for i in range(num_iterations):
    grad = grad_f(x)
    x = x - learning_rate * grad
    x_values.append(x)
    f_values.append(f(x))

# Visualización
x_plot = np.linspace(-1.5, 3.5, 400)
f_plot = f(x_plot)

plt.figure(figsize=(8, 6))
plt.plot(x_plot, f_plot, label='$f(x) = x^4 - x^2$')
plt.plot(x_values, f_values, 'ro-', label='Descenso de Gradiente')
plt.title('Descenso de Gradiente con Mínimo Local')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Función y gradiente
def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    return np.array([2*x, 2*y])

# Parámetros iniciales
point = np.array([5.0, 5.0])
learning_rate = 0.1
num_iterations = 50
points = [point.copy()]
f_values = [f(point[0], point[1])]

for i in range(num_iterations):
    grad = grad_f(point[0], point[1])
    point = point - learning_rate * grad
    points.append(point.copy())
    f_values.append(f(point[0], point[1]))

# Preparación de datos para la gráfica
X, Y = np.meshgrid(np.linspace(-6, 6, 100), np.linspace(-6, 6, 100))
Z = f(X, Y)
points = np.array(points)

# Visualización
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.5)
ax.plot(points[:,0], points[:,1], f_values, 'r.-')
ax.set_title('Descenso de Gradiente en 3D')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Definir la función y su gradiente
def f(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 -7)**2

def grad_f(x, y):
    df_dx = 4*(x**2 + y - 11)*x + 2*(x + y**2 -7)
    df_dy = 2*(x**2 + y - 11) + 4*(x + y**2 -7)*y
    return np.array([df_dx, df_dy])

# Parámetros iniciales
point = np.array([2.0, -5.0])  # Cambia este punto para observar diferentes comportamientos
learning_rate = 0.001
num_iterations = 2000
points = [point.copy()]
f_values = [f(point[0], point[1])]

# Algoritmo de Descenso de Gradiente
for i in range(num_iterations):
    grad = grad_f(point[0], point[1])
    point = point - learning_rate * grad
    points.append(point.copy())
    f_values.append(f(point[0], point[1]))

# Preparación de datos para la gráfica
X = np.linspace(-6, 6, 400)
Y = np.linspace(-6, 6, 400)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
points = np.array(points)

# Visualización
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Superficie de la función
ax.plot_surface(X, Y, Z, cmap=cm.viridis, alpha=0.7, linewidth=0, antialiased=False)

# Trayectoria del descenso de gradiente
ax.plot(points[:,0], points[:,1], f_values, 'r.-', label='Descenso de Gradiente', markersize=5)

# Configuraciones adicionales
ax.set_title('Descenso de Gradiente en 3D sobre la Función de Himmelblau')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
ax.view_init(45, 45)  # Ángulo de visualización
ax.legend()
plt.show()
```

```python
import torch

# Datos de entrenamiento
Celsius = torch.tensor([[40], [10], [0], [8], [15], [22], [38]], dtype=torch.float32)
Fahrenheit = Celsius * 9/5 + 32

# Normalizar Fahrenheit
Celsius = (Celsius - Celsius.min()) / (Celsius.max() - Celsius.min())
Fahrenheit = (Fahrenheit - Fahrenheit.min()) / (Fahrenheit.max() - Fahrenheit.min())

# Reentrenar el modelo con datos normalizados


import torch.nn as nn

# Modelo del Perceptrón
model = nn.Sequential(
    nn.Linear(1, 1),  # Una neurona
    nn.Sigmoid()      # Función de activación sigmoide
)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
loss_values = []
weights = []
biases = []

for epoch in range(1000):
    optimizer.zero_grad()
    outputs = model(Celsius)
    loss = criterion(outputs, Fahrenheit)
    loss.backward()
    optimizer.step()

    loss_values.append(loss.item())

    # Almacenar pesos y sesgos para visualización
    [w, b] = model[0].parameters()
    weights.append(w.item())
    biases.append(b.item())
#    print([w, b])
```

```python
import numpy as np
import plotly.graph_objects as go


# Crear malla de valores para peso y sesgo
w_range = np.linspace(-15.0, 15.0, 50)
b_range = np.linspace(-15.0, 15.0, 50)
W, B = np.meshgrid(w_range, b_range)
Loss_surface = np.zeros_like(W)

# Calcular pérdida para cada combinación de w y b
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        y_pred = torch.sigmoid(Celsius * W[i,j] + B[i,j])
        #y_pred = (Celsius * W[i,j] + B[i,j])
        loss = criterion(y_pred, Fahrenheit)
        Loss_surface[i,j] = loss.item()

# Convertir listas de pesos y sesgos a arrays
weights = np.array(weights)
biases = np.array(biases)

# Crear figura interactiva con plotly
fig = go.Figure(data=[
    go.Surface(z=Loss_surface, x=W, y=B, colorscale='Viridis', opacity=0.8),
    go.Scatter3d(x=weights, y=biases, z=loss_values, mode='lines+markers',
                 line=dict(color='red', width=4), marker=dict(size=5),
                 name='Trayectoria de entrenamiento')
])

# Configurar etiquetas y título
fig.update_layout(
    scene=dict(
        xaxis_title='Peso',
        yaxis_title='Sesgo',
        zaxis_title='Pérdida'
    ),
    title='Superficie de Error y Descenso de Gradiente',
)

# Mostrar figura interactiva
fig.show()
```

```python
# Generación de datos
x_train = torch.linspace(-4*np.pi, 4*np.pi, 10000).unsqueeze(1)
y_train = torch.sin(x_train)
```

```python
# Modelo con una capa oculta
model = nn.Sequential(
    nn.Linear(1, 20),  # Capa oculta con 10 neuronas
    nn.ReLU(),         # Función de activación
    nn.Linear(20, 20),  # Capa oculta con 10 neuronas
    nn.Tanh(),
    nn.Linear(20, 1),   # Capa de salida
    nn.Tanh()
)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

loss_values = []

for epoch in range(8000):
    optimizer.zero_grad()
    outputs = model(x_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()

    loss_values.append(loss.item())
```

```python
# Predicciones
with torch.no_grad():
    predictions = model(x_train)

# Gráfica
plt.plot(x_train.numpy(), y_train.numpy(), label='Real')
plt.plot(x_train.numpy(), predictions.numpy(), label='Predicción')
plt.legend()
plt.title('Ajuste de la Red Neuronal a y = sin(x)')
plt.show()
```
