```python
import torch

# Definir los vectores
a = torch.tensor([1, 3, -5], dtype=torch.float32)
b = torch.tensor([4, -2, -1], dtype=torch.float32)

# Calcular el producto punto
producto_punto = torch.dot(a, b)
print(f"Producto punto de {a.numpy()} y {b.numpy()}: {producto_punto.item()}")
```

```python
# Definir los vectores
a = torch.tensor([1, 2], dtype=torch.float32)
b = torch.tensor([3, 4, 5], dtype=torch.float32)

# Calcular el producto externo
producto_externo = torch.ger(a, b)
print(f"Producto externo de {a.numpy()} y {b.numpy()}:\n{producto_externo}")
```

```python
# Definir los vectores
a = torch.tensor([1, 0, 0], dtype=torch.float32)
b = torch.tensor([0, 1, 0], dtype=torch.float32)

# Calcular el producto cruz
producto_cruz = torch.cross(a, b)
print(f"Producto cruz de {a.numpy()} y {b.numpy()}: {producto_cruz.numpy()}")
```

```python
# Definir las matrices
A = torch.tensor([[1, 2],
                  [3, 4]], dtype=torch.float32)
B = torch.tensor([[5, 6],
                  [7, 8]], dtype=torch.float32)

# Multiplicación de matrices
C = torch.matmul(A, B)
print(f"Matriz A:\n{A}")
print(f"Matriz B:\n{B}")
print(f"Producto matricial A x B:\n{C}")
```

```python
# Multiplicación A x B
C = torch.matmul(A, B)

# Multiplicación B x A
D = torch.matmul(B, A)

print(f"A x B:\n{C}")
print(f"B x A:\n{D}")
```

```python
# Producto elemento a elemento
E = A * B
print(f"Producto elemento a elemento A * B:\n{E}")
```

```python
# Crear tensores aleatorios
T = torch.rand(2, 3, 4)
S = torch.rand(4, 5, 6)

# Contracción sobre la tercera dimensión de T y la primera de S
R = torch.tensordot(T, S, dims=([2], [0]))
print(f"Dimensiones de T: {T.shape}")
print(f"Dimensiones de S: {S.shape}")
print(f"Dimensiones del resultado R: {R.shape}")
```

```python
# Definir los tensores
A = torch.tensor([1, 2], dtype=torch.float32)
B = torch.tensor([3, 4, 5], dtype=torch.float32)

# Producto tensorial
C = torch.ger(A, B)
print(f"Producto tensorial de A y B:\n{C}")
```

```python
# Definir las matrices
A = torch.tensor([[1, 2],
                  [3, 4]], dtype=torch.float32)
B = torch.tensor([[5, 6],
                  [7, 8]], dtype=torch.float32)

# Multiplicación de matrices usando einsum
C = torch.einsum('ij,jk->ik', A, B)
print(f"Resultado usando einsum:\n{C}")
```

```python
# Simular una capa densa
entrada = torch.rand(1, 784)  # Vector de entrada (ejemplo: imagen aplanada de 28x28 píxeles)
pesos = torch.rand(784, 128)  # Pesos de la capa
sesgos = torch.rand(128)      # Sesgos de la capa

# Calcular la salida
salida = torch.matmul(entrada, pesos) + sesgos
print(f"Dimensiones de la salida: {salida.shape}")
```
