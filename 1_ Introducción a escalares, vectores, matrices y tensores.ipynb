```python
import torch
import matplotlib.pyplot as plt

# Definir un vector en R^2
vector = torch.tensor([2, 3], dtype=torch.float32)
print(f"Vector: {vector}")

# Visualización del vector en 2D
plt.figure()
plt.quiver(0, 0, vector[0],
 vector[1], angles='xy', scale_units='xy', scale=1,color='red')
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Visualización de un Vector en 2D')




plt.grid()
plt.show()
```

```python
import torch

# Definir un escalar
escalar = torch.tensor(5)
print(f"Escalar: {escalar}")
```

```python

# Definir dos vectores
vector_a = torch.tensor([1, 2], dtype=torch.float32)
vector_b = torch.tensor([3, 1], dtype=torch.float32)

# Suma de vectores
vector_suma = vector_a + vector_b
print(f"Suma de vectores: {vector_suma}")

# Visualización de la suma de vectores
plt.figure()
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector A')
plt.quiver(vector_a[0], vector_a[1], vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector B')
plt.quiver(0, 0, vector_suma[0], vector_suma[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A + B')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Suma de Vectores')
plt.legend()
plt.grid()
plt.show()
```

```python
import torch

# Crear un tensor 3x3x3 con valores aleatorios
tensor = torch.rand(3, 3, 3)
print(f"Tensor de ejemplo (3x3x3):\n{tensor}")
```

```python
# Acceder al primer elemento del tensor
elemento = tensor[0, 0, 0]
print(f"Primer elemento del tensor: {elemento}")

# Extraer un subtensor
subtensor = tensor[0, :, :]
print(f"Subtensor (primera matriz 2D del tensor 3D):\n{subtensor}")
```

```python
# Sumar un escalar al tensor
tensor_suma = tensor + 5
print(f"Tensor después de sumar 5 a cada elemento:\n{tensor_suma}")
```

```python
import torch

# Definir un vector
vector = torch.tensor([3.0, 4.0])

# Calcular la norma L2 (Euclidiana)
norma_l2 = torch.norm(vector, p=2)
print(f"Norma L2 del vector {vector.numpy()}: {norma_l2.item()}")

# Calcular la norma L1
norma_l1 = torch.norm(vector, p=1)
print(f"Norma L1 del vector {vector.numpy()}: {norma_l1.item()}")
```

```python
import torch

# Definir dos vectores
u = torch.tensor([1.0, 2.0, 3.0])
v = torch.tensor([4.0, 5.0, 6.0])

# Calcular el producto punto
producto_punto = torch.dot(u, v)
print(f"Producto punto de {u.numpy()} y {v.numpy()}: {producto_punto.item()}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir vectores en R^2 para visualización
u = torch.tensor([1.0, 0.0])
v = torch.tensor([0.2, 0.5])

# Calcular el producto punto
producto_punto = torch.dot(u, v)
print(f"Producto punto de {u.numpy()} y {v.numpy()}: {producto_punto.item()}")

# Visualización
plt.figure()
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector v')
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Vectores Ortogonales en R^2')
plt.legend()
plt.grid()
plt.show()
```

```python
# Definir matrices
A = torch.tensor([[1.0, 2.0],
                  [3.0, 4.0]])

B = torch.tensor([[5.0, 6.0],
                  [7.0, 8.0]])

# Multiplicación de matrices
C = torch.matmul(A, B)
print(f"Producto matricial A x B:\n{C}")
```

```python
import torch

# Supongamos que tenemos una imagen en escala de grises de 28x28 píxeles
imagen = torch.rand(28, 28)
print(f"Dimensiones de la imagen: {imagen.shape}")

# Para un lote de 64 imágenes
batch_imagenes = torch.rand(64, 28, 28)
print(f"Dimensiones del lote de imágenes: {batch_imagenes.shape}")
```

```python
# Definir pesos y sesgos de una capa densa (fully connected)
# Supongamos una capa con 784 entradas y 128 neuronas
pesos = torch.rand(784, 128)
sesgos = torch.rand(128)

print(f"Dimensiones de los pesos: {pesos.shape}")
print(f"Dimensiones de los sesgos: {sesgos.shape}")
```

```python
# Entrada de la red (por ejemplo, una imagen aplanada de 28x28 píxeles)
entrada = torch.rand(1, 784)  # Batch size de 1

# Operación de la capa (entrada x pesos + sesgos)
salida = torch.matmul(entrada, pesos) + sesgos
print(f"Salida de la capa: {salida.shape}")
```

```python
import torch
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir un vector en R^3
vector_3d = torch.tensor([2, 3, 4], dtype=torch.float32)
print(f"Vector en 3D: {vector_3d}")

# Visualización del vector en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, vector_3d[0], vector_3d[1], vector_3d[2], length=1, normalize=True, color='r')
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Visualización de un Vector en 3D')
plt.show()
```
